```python
import numpy as np
import pylab as py

mylines = np.loadtxt('Hubbard.dat', float)
#print(mylines)

temp = mylines[0:50:1,1]
#print(temp)

chem = mylines[0:851:1,0]
#print(chem)

density = mylines[0::1,2]
#print(density)

energy = mylines[0::1,3]
#print(energy)

density = density.reshape(851,50)
#print(density)

energy = energy.reshape(851,50)
#print(energy)


fig1 = py.figure(figsize=(10,10));
py.title('Energy as function of Temperature for 426th mu', fontsize=25)
py.grid() #turns grid on the graph on 
py.xlabel('Temperature for 426th mu', fontsize=25)
py.ylabel('Energy', fontsize=25)
py.rcParams.update({'font.size': 20})
py.plot(temp, energy[426,0:50:1])

fig2 = py.figure(figsize=(10,10));
py.title('Density vs Temperature', fontsize=25)
py.grid()
py.xlabel('Temperature for 426th mu', fontsize=25)
py.ylabel('Density', fontsize=25)
py.rcParams.update({'font.size': 20})
py.plot(temp, density[426,0:50:1] )

#Part G in next cell
```

```python
def get_mu_index(rho):
    mu_index = np.argmin(np.abs(density - rho), axis = 0) 
    return mu_index

x2 = np.linspace(.65, .95, 7)


fig3 = py.figure(figsize=(10,10));

    
for p in x2:
    E = np.zeros(50)
    mu_idx = get_mu_index(p)

    for i in range(50):
        E[i] = energy[mu_idx[i],i]
    py.plot(temp, E, label = round(p,2))

py.plot(temp, E, label = "0.9")
py.legend(loc='best',fontsize=15)
```
