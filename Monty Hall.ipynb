```python
import numpy as np
import random as rand
import pylab as py

N = 3 #Number of doors

Zonk = 2

KeepDoor = []
SwitchDoor = []

Doors = np.zeros(N)

RD = rand.randint(0,N-1)
Doors[RD] = 1

for A in range(100):

    Choice = rand.randint(0, N-1)

    
#These conditions aren't really needed/more for manual inputs etc...
    for i in range(len(Doors)):
        if i == Choice:
            pass
        elif i == Zonk:
            pass
        elif i == RD:
            pass
        else:
            #print(f'Door Number {i+1} has a zonk')
            Doors[i] = Zonk

    if Choice == RD:
        KeepDoor.append(1)

    else:
        SwitchDoor.append(1)

#print(Doors)
```

```python
fig1 = py.figure(figsize=(8,8));
py.rcParams.update({'font.size': 15})
x = [0,10]
y = [len(SwitchDoor),len(KeepDoor)]

py.title('Monty Hall')
py.xlabel('Match Outcome', fontsize=15)
py.ylabel('Number of Wins', fontsize=15)
py.xticks([])

CC= np.array(['red','blue'])
LL= np.array(['Switch Door Wins', 'Keep Door Wins'])

for P in range(2):
    py.bar(x[P],y[P],5,align='center', label=LL[P],color=CC[P])
    py.annotate(y[P], xy=(x[P], y[P]+ 0.001), fontsize=15, va='bottom', ha='center')
    
py.legend(loc='best',fontsize = 15)
```
