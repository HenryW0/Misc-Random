```python
from pylab import *

N=eval(input('Enter the desired fractional error: ')) #Can be used to change fractional error, actual value will be less

def user_input(N):
    if N <= 0:
        raise ValueError('Fractional Error must be between 0 and 1')
    elif N >= 1:
        raise ValueError('Fractional Error must be between 0 and 1')
    return user_input
UI=user_input(N)

print('') #Space between outputs
 
def Euler_Pi():
    x, sum1 = 1,0
    while x < 1000000 : #Capped at 1 million terms
        sum1 += (6/(x**2))
        Euler_Error = abs(((pi)-sqrt(sum1))/(pi))
        if Euler_Error <= N: #10
            print('Euler\'s Approximation is %.16f' %(sqrt(sum1)))
            print('Minimum amount of terms needed for fractional error less than %e: %d' %(N,x))
            print('The fractional error for Euler\'s Approximation is %e' %(Euler_Error))
            break
        x += 1
    return Euler_Pi

Pi_E=Euler_Pi()
#For the Euler's Approximation, a minimum of 303,964 terms are needed to converge to a fractional error of 10^-6 or less

print('') #Space between outputs

def Madhava_Pi() :
    x, sum2 = 1,0
    while x < 1000000 : #Capped at 1 million terms
        sum2 += ((-1)**(x+1))/((2*x-1)*(3**(x-1)))
        Madhava_Error = abs((pi-sqrt(12)*(sum2))/pi)
        if Madhava_Error <= N: #10
            print('Madhava\'s Approximation is %.16f' %(sqrt(12)*sum2))
            print('Minimum amount of terms needed for fractional error less than %e: %d' %(N,x))
            print('The fractional error for Madhava\'s Approximation is %e' %(Madhava_Error))
            break
        x += 1
    return Madhava_Pi

Pi_M=Madhava_Pi()
#For the Madhava's Approximation, a minimum of 10 terms are needed to converge to a fractional error of 10^-6 or less



F=array(arange(1,1001,2)) #Reasonable amount of N_max terms
F2=array(arange(1,4,2)) #Unreasonable amount of N_max terms
x1=array(arange(1,51,1))

def Fourier_Series():
    Fourier_Series=zeros(len(x1))
    for x in range(50):
        Fourier = 0.5
        for n in F :
            Fourier += ((2/(pi*n)) * sin(n*x))
        Fourier_Series[x]= Fourier
    return Fourier_Series
    
FS=Fourier_Series()

def Fourier_Series2():
    Fourier_Series2=zeros(len(x1))
    for x in range(50):
        Fourier2 = 0.5
        for n in F2 :
            Fourier2 += ((2/(pi*n)) * sin(n*x))
        Fourier_Series2[x]= Fourier2
    return Fourier_Series2
    
FS2=Fourier_Series2()

title('Square Wave Fourier Series', fontsize=25)
xlabel('x-axis', fontsize=25)
ylabel('y-axis', fontsize=25)
#grid(color='black', linestyle='-', linewidth=.5) #Makes graph look too messy
rcParams['figure.figsize'] = 10,10
rcParams.update({'font.size': 20})
xticks(arange(0.0, 51, step=5))
yticks(arange(0, 1.1, step=0.1))
plot(x1,FS,label='Reasonable',color='green',linestyle='-',linewidth=2)
plot(x1,FS2,label='Unreasonable',color='red',linestyle='--',linewidth=1.75)
legend(loc='best',fontsize=15)
#savefig('Henry_Wahhab_Homework4_SquareWave.png', bbox_inches = 'tight')
```
