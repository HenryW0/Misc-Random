```python
from pylab import *
from ipywidgets import interact 
from time import time

#FORWARD EULER

def slope(y):
    slope = y #INSERT dy/dx as a function in terms of y 
    return slope

value=float(input('Enter x value needed: '))
step=float(input('Enter step size: '))
y_i=float(input('Enter initial y value: '))

rep=int(round(value/step))

y_val=zeros(rep+1)
y_val[0]=y_i

time_start=time()

for n in range(rep):
    y_val[n+1]=y_val[n]+(step)*(slope(y_val[n])) #y'=4y-y^2
    
time_final=time()

print('The value at %.3f is approximately %.10f' %(value,y_val[rep]))


def widget(print_data):
    if print_data == True:
        print('%.d iterations' %(rep))
        print('Computational time: %.5f seconds' %(time_final-time_start))
        print(y_val)
    else:
        print()
    
interact(widget, print_data=False)
show()

#3.523188311911529776239
#3.523188312280406 Closest estimation, 1B repetitions, ~31 min comp time
```

```python
from pylab import *

#IMPROVED EULER METHOD CODE TEMPLATE f'(x,y)= x + 2cos(xy)

value=float(input('Enter x value needed: '))
h=float(input('Enter step size: '))
y_i=float(input('Enter initial y value: '))

iterations=len(linspace(0,value,int(value/h)))

def foo(x,y):
    foo = 4*cos(x*y) + x #slope function y' 
    return foo


x=zeros(iterations+2) 
y=zeros(iterations+2)
y[0]=y_i

def goo(x,y):
    goo = (y+h*(foo(x,y))) #initial y + (x increment times slope)
    return goo
    
for i in range(iterations+1):
    x[i+1] = x[i] + h
    y[i+1]= y[i] + (h*(foo(x[i],y[i])+ foo(x[i+1],goo(x[i],y[i]))))/2
    print(round(x[i],2),round(y[i],5))


fig1 = figure(figsize=(10,10));
grid(color='black', linestyle='-', linewidth=.5)
plot(x,y,color='red')
```
