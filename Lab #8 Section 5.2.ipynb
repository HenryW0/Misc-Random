```python
from pylab import *

def newarctan(x,y) :
    if x > 0 and y > 0 :
        newarctan=(arctan(y/x) * (180/pi))
        return newarctan
    elif x < 0 and y > 0 :
        newarctan=((arctan(y/x) + (pi)) * (180/pi)) 
        return newarctan
    elif x < 0 and y < 0 :
        newarctan=((arctan(y/x) + (pi)) * (180/pi))
        return newarctan
    elif x > 0 and y < 0 :
        newarctan=((arctan(y/x) +(2*pi)) * (180/pi))
        return newarctan
    elif x==0 and y > 0 :
        newarctan=90
        return newarctan
    elif x==0 and y < 0 :
        newarctan=270
        return newarctan
    elif x > 0 and y==0 :
        newarctan=360 #0 degrees = 360 degrees
        return newarctan
    elif x < 0 and y==0 :
        newarctan=180
        return newarctan
    else:
        newarctan=0
        return newarctan
(x,y) = eval(input('Enter Point (x,y): '))
z = newarctan(x,y)
print('The angle between the positive X-Axis and the point is %.3f degrees' %(z))


x1=arange(-10,11,1)
def angles():
    angles=zeros((len(x1)))
    for i in range (-10,11):
            t=round(newarctan(x1[i],5)) #Y is a fixed value of 5
            angles[i] = t
    return angles

#Graph in the next cell
```

```python
x2=linspace(-10,10,100)
x3=linspace(-10,10,50)
C=arctan(5/x2)*(180/pi) #Y is a fixed value of 5
D=arctan2(5,x3)*(180/pi)#Y is a fixed value of 5
B=angles()
plot(x1,B,label='newarctan',linewidth=6,color='green')
plot(x2,C,label='arctan',color='red',linestyle='-.', linewidth=4)
plot(x3,D,'*',label='arctan2',color='black', linewidth=3)
rcParams.update({'font.size': 20})
rcParams['figure.figsize'] = 12,12
grid(color='black', linestyle='-', linewidth=.5)
title('Angles vs. X-Values (with a fixed Y-Value of 5)', fontsize=25)
xlabel('x-axis', fontsize=25)
ylabel('Angles $(Degrees)$', fontsize=25)
ylim(-100,160)
yticks(range(-100,160,25))
legend(loc='best',fontsize=30)
```
