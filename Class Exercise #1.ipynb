```python
from pylab import *
import numpy as np

def f(x):
    x0 = 0
    k = 3.0
    sigma = 1.0
    return np.cos(k*(x-x0))*np.exp(-(x-x0)**2/2/sigma**2)
```

```python
Nx = 200 #Grid points
Nt = 1000 #Time grid
v = 1.0
h = 0.1
x = np.linspace(-10,10, Nx)
a = x[1] - x[0]

C = v * h/2/a
print(C)
```

```python
A = np.zeros([Nx, Nt])
A[:,0] = f(x)

for n in range(0, Nt-1): 
    for i in range(Nx):
        A[i, n+1] = 0.5*(A[i-1,n] + A[(i+1)%Nx, n]) - C* (A[(i+1)%Nx, n] - A[i-1, n])
```

```python
plot(x,f(x), color = 'red', label = 'Original')
plot(x,A[:,50], color = 'blue', label = 'Moving')
legend()
show()
```
