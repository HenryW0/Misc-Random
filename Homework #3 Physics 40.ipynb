```python
from pylab import *

x1=linspace(0.0,1.0,100)
y1=x1
y2=x1**2
fig1 = figure(figsize=(10,10));
plot(x1,y1,color='red',linewidth=2,label='y=x')
plot(x1,y2,color='blue',linewidth=2,label='y=x')
title('Graph 1: y=x VS. y=x', fontsize=25)
xlabel('x-axis', fontsize=25)
ylabel('y-axis', fontsize=25)
grid(color='black', linestyle='-', linewidth=.5)
xticks(arange(0.0, 1.1, step=0.1))
yticks(arange(0.0, 1.1, step=0.1))
legend(loc='best',fontsize =30)

fig2 = figure(figsize=(10,10));
plot(x1,log(y1),color='green',linewidth=2,label='y=log(x)')
plot(x1,log(y2),color='magenta',linewidth=2,label='y=log(x)')
title('Graph 2: Log()', fontsize=25)
xlabel('x-axis', fontsize=25)
ylabel('y-axis', fontsize=25)
grid(color='black', linestyle='-', linewidth=.5)
xticks(arange(0.0, 1.1, step=0.1))
yticks(arange(0.0, -10, step=-1))
legend(loc='best',fontsize =30)
#When the logarithm is equal to 0, an error occurs because the graph is undefined at that point because any number to the power of
#another number cannot be equal to 0, in this case 10**n=0

fig3 = figure(figsize=(10,10));
loglog(x1,y1,label='loglog(x)',color='orange',linewidth=2)
loglog(x1,y2,label='loglog(x)',color='cyan',linewidth=2)
title('Graph 3: loglog()', fontsize=25)
xlabel('x-axis', fontsize=25)
ylabel('y-axis', fontsize=25)
grid(color='black', linestyle='-', linewidth=.5)
legend(loc='best',fontsize =30)

#Graphs may glitch out the first time when enterring an input, please retry to see the correctly scaled graph
```

```python
#Two small spheres spaced (insert value) meters apart have equal charge.
#How many excess electrons must be present in each sphere if the magnitude of the force between them is (insert value) newtons?
from pylab import *
e_charge, k = 1.6*10**-19, 9.0*10**9
def electrons(Force, Distance):
    if F < 0 :
        electrons = sqrt((abs(F)*r**2)/k)
        return electrons
    else:
        electrons=sqrt((F*r**2)/k)
        return electrons
F,r= eval(input('Enter Force of attraction/repulsion (Newtons) and distance (meters): '))
q=electrons(F,r)
n=round(q/e_charge)
print('There are %d excess electrons present in each sphere' %(n))

x1=linspace(0,10**-21,100)
Q=sqrt((x1*.25**2)/k) #Assuming Distance between the particles is 0.25 meters
n2=(Q/e_charge)
plot(x1,n2,'o',color='red', linewidth=2) #Used 'o' to show that the graph is not continuous because the number of electrons can only be whole numbers
grid(color='black', linestyle='-', linewidth=.5)
title('Number of Electrons VS: Force at a fixed distance ', fontsize=25)
xlabel('$Force$ [Newtons]', fontsize=25)
ylabel('$Electrons$', fontsize=25)
xticks(arange(0.0,9**-22,step=10**-22))
yticks(arange(0.0, 526, step=25))
text(.86*10**-21,-57,'X-Scale:',fontsize=18, color='magenta')
rcParams.update({'font.size': 20})
rcParams['figure.figsize'] = 15,15
```
