```python
from math import sin,cos,pi

# Ask the user for the values of r and theta
r = float(input("Enter r: "))
d = float(input("Enter theta in degrees: "))

# Convert the angle to radians
theta = d*pi/180

# Calculate the equivalent Cartesian coordinates
x = r*cos(theta)
y = r*sin(theta)

# Print out the results
print("x = ",x,", y = ",y)
```

```python
from math import atan, sqrt, pi

#Ask the user for values of x and y 
x = float(input("Enter x: "))
y = float(input("Enter y: "))

#Obtain angle and convert to degrees

d = round(atan(y/x) * 180/pi ,2) #theta

#Calculate the equivalent radius 

r = round(sqrt(x**2 + y**2),2)

# Print out the results

print('Polar Coordinates:')
print('r =',r, " d =",str(d) + " degrees")
```

```python
from matplotlib.pylab import *
x = (np.linspace(-10, 10, 1000)) #creates an array with 1000 terms evenly divided 

def Fourier(terms, x): #10 terms
    Fourier = zeros(len(x))
    for n in range(1, terms*2, 2) :
        Fourier += ((2/(pi*n)) * sin(n*x))
    return Fourier


#Fix plot and make it more readable
title('Square Wave Fourier Series', fontsize=25)
xlabel('x-axis', fontsize=25)
ylabel('y-axis', fontsize=25)
rcParams['figure.figsize'] = 10,10
rcParams.update({'font.size': 20})

plt.plot(x, Fourier(2, x) + 0.5 , label='2 Terms') #2 terms
plt.plot(x, Fourier(10, x) + 0.5, label='10 Terms') #10 terms
plt.plot(x, Fourier(100,x) + 0.5 , label='100 Terms') #100 terms
plt.plot(x, (sign(sin(x)) + 1)/2 , label='Exact Function') #exact function
legend(loc='best',fontsize=15)
plt.show()
```
