```python
from pylab import *

b, c, d, e, f, g ,h, i = 7.0081,8.7194,59.7367,91.9391,124.9144,210.6070,331.6006,14.4491
Planets=array([b,c,d,e,f,g,h,i])/365
name=array(['b','c','d','e','f','g','h','i'])

def Semi(P):
    Semi=(P**2)**(1/3)
    return Semi

for i in range(8):
    print('planet %s : semi-major axis = %.5f AU = %.3f km' %(name[i],Semi(Planets[i]),Semi(Planets[i])*1.5*10**8))

    
R=array([1.31,1.18,2.88,2.67,2.89,8.13,11.32,1.32])
n=100
theta=linspace(0,2*pi,n)
#Testing

def X(x_0,R,n):
    X= R*cos(theta) + x_0
    return X

def Y(y_0,R,n):
    Y= R*sin(theta) + y_0
    return Y

def plotting():
    for j in range(8):
        plot(X(0,Semi(Planets[j]),n),Y(0,Semi(Planets[j]),n),linestyle='--',linewidth=2,color='black')
        plot(X(Semi(Planets[j]),R[j]/130,n),Y(0,R[j]/130,n),linewidth=2,color='black')
    return plotting
    
    
fig1 = figure(figsize=(15,15));
title(' Kepler-90 Exoplanet Orbit', fontsize=25)
xlabel('Horiztonal Distance from Orbit Center [AU]', fontsize=20)
ylabel('Vertical Distance from Orbit Center [AU]', fontsize=20)
grid(color='black', linestyle='-', linewidth=.5)
rcParams.update({'font.size': 20})
plot(X(0,.05,n),Y(0,.05,n),linestyle='--',linewidth=2,color='red')
Circle=plotting()
#savefig('Midterm_OrbitFig.png')

#Moore's Law Section in next Cell
```

```python
from pylab import *

Years=array([1972,1974,1978,1982,1985,1989,1993,1997,1999,2000,2003,2004,2007,2008,2012])
TS=array([.0025,.005,0.029,.12,.275,1.18,3.1,7.5,24.0,42.0,220.0,592.0,1720.0,2046.0,3100.0])*1000000

def Moore(n,y_i,y_0):
    Moore= log10(n) + ((y_i-y_0)/2) * log10(2)
    return Moore

def plottingM():
    for k in range(15):
        plot(Years[k],Moore(TS[k],Years[k],1971))
    return plottingM

fig1 = figure(figsize=(12,12));
title("Moore's Law", fontsize=25)
xlabel('Year', fontsize=20)
ylabel('log(Transistors)', fontsize=20)
grid(color='black', linestyle='-', linewidth=.5)
rcParams.update({'font.size': 20})    
ylim(3.1,10)
xlim(1970,2015)
plot(Years,log10(TS),'*',markersize=15,color='red')
M=plottingM()

#Not completed
```
