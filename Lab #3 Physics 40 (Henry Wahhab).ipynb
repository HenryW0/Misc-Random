```python
from pylab import *
x=linspace(0.0, 12, 100) #range, increments
y=2*x**2
y2=4*x
plot(x, y, '--', label ='blue line')
plot(x, y2, label ='orange line')
x2= linspace(0.0,12,10)
y3=x2+250
plot(x2, y3,'o', label= 'green circle')
plot(x2,y3+50, 'x', label ='red X')
plot(x2,y3-50,'*', label = 'purple star')
y4=4+(x*0) #use axvline() or axhline() for vertical and horizontal lines
plot(x, y4, 'c-.', label = 'cyan line')
xlabel('x axis', fontsize=20)
ylabel('y axis', fontsize=20)
title('Lab 3 Graph',fontsize=20)
text(4, 150, 'label', fontsize=14, color='magenta')
plt.axis('on')
grid(color='r', linestyle='-', linewidth=1)
legend(loc='lower right', fontsize =12)
rcParams.update({'font.size': 15})
plt.xkcd(0,0,0) #There are 3 values for this command, 1. Scale 2. Length 3. Randomness, each affect the style and font of the graph
#savefig('Henry_Wahhab_lab3graph.png', bbox_inches = 'tight')
```
