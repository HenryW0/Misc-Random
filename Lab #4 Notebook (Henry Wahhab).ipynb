```python
from pylab import *
def position(time) :
    position = (-0.2*t**3) + (0.5*t**2) + 4*t - 22
    return position 
t = eval(input('Enter time in seconds: '))
x = position(t)
print('The position (x) is %.3f m at time %.3f [s]' %(x,t))


def velocity(time) :
    velocity = (-0.6*t**2) + 1.0*t + 4
    return velocity
v = velocity(t)
print('The velocity (v) is %.3f m/s at time %.3f [s]' %(v,t))

def acceleration(time) :
    acceleration = -1.2*t + 1.0
    return acceleration
a= acceleration(t)
print('The acceleration (a) is %.3f m/s^2 at time %.3f [s]' %(a,t))

x1= linspace(0.0,8.0,48)

subplot(2,2,1);
y1= (-0.2*x1**3) + (0.5*x1**2) + 4*x1 - 22;
plot(x1,y1, color='blue')
title('Position vs Time Graph $x(t)$', fontsize=25)
xlabel('$Time$ [s]')
ylabel('$Position$ [m]')
grid(color='black', linestyle='-', linewidth=.5)


subplot(2,2,2);
y2=(-0.6*x1**2) + 1.0*x1 + 4;
plot(x1,y2,color='orange')
title('Velocity vs Time Graph $v(t)$', fontsize=25)
xlabel('$Time$ [s]')
ylabel('$Velocity$ [m/s]')
grid(color='black', linestyle='-', linewidth=.5)

subplot(2,2,(3,4));
y3= (-1.2*x1) + 1.0;
plot(x1,y3,'r')
title('Acceleration vs Time Graph $a(t)$', fontsize=25)
xlabel('$Time$ [s]')
ylabel('$Acceleration$ [m/s^2]')
grid(color='black', linestyle='-', linewidth=.5)

subplots_adjust(hspace=.4, wspace=.4)
rcParams.update({'font.size': 17})
rcParams['figure.figsize'] = 20, 20
#savefig('Henry_Wahhab_Lab4Graphs.png', bbox_inches = 'tight') #Remove pound sign to save image
```

```python
from pylab import *
#Works for real numbers only
def quadratic1(a , b , c) : 
    quadratic1 = (-b+sqrt(b**2-4*a*c))/(2*a)
    return quadratic1
a,b,c = eval(input('Enter coefficients (a,b,c): '))
x1=quadratic1(a, b, c)

def quadratic2(a , b , c) : 
    quadratic2 = (-b-sqrt(b**2-4*a*c))/(2*a)
    return quadratic2
x2=quadratic2(a, b, c)
print('x is equal to %.3f and %.3f' %(x1, x2))
```
