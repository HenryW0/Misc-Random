```python
from pylab import *

h=6.626 * 10**-34
c=2.99 * 10**8
k= 1.381 * 10**-23

def blackbody(L,T):
    blackbody=((2 * h * (c**2))/((L*10**-9)**5))*(1/(e**((h*c)/((L*10**-9)*k*T))-1))
    return blackbody
L,T=eval(input('Enter Wavelength [nm] and Temperature [K]: '))
BB=blackbody(L,T)
print('The Black Body Radiation is %.3e [W/m^3/sterad]' %(BB))

T1=array(arange(4000,7001,1000))

def plotting():
    i=0
    while i < 4:
        x2=linspace(100,1000,100)
        y2=((2 * h * (c**2))/((x1*10**-9)**5))*(1/(e**((h*c)/((x2*10**-9)*k*T1[i]))-1))
        plot(x2,y2,label='%d Kelvin' %(T1[i]))
        i += 1
        if i==2 :
            i += 1
            pass
        elif i==3 :
            x2=linspace(100,1000,100)
            y2=((2 * h * (c**2))/((x1*10**-9)**5))*(1/(e**((h*c)/((x2*10**-9)*k*T1[i]))-1))
            plot(x2,y2,label='7000 Kelvin')
            break
            

fig1 = figure(figsize=(12,12));
x1=linspace(100,1000,100)
y1=((2 * h * (c**2))/((x1*10**-9)**5))*(1/(e**((h*c)/((x1*10**-9)*k*6000))-1))
xlim(100,1000)
Graph=plotting()
title('Blackbody Radiation', fontsize=25)
xlabel('$Wavelength$ [nm]', fontsize=25)
ylabel('$B(,T)$ [W/m^3/sterad])', fontsize=25)
grid(color='black', linestyle='-', linewidth=.5)
plot(x1,y1,label='6000 Kelvin')

legend(loc='best',fontsize=20)
rcParams.update({'font.size': 20})


lambdaMax=(0.002897/T1)*(10**9)

plot(lambdaMax[0],blackbody(lambdaMax[0],4000),'o',color='blue')
plot(lambdaMax[1],blackbody(lambdaMax[1],5000),'o',color='orange')
plot(lambdaMax[2],blackbody(lambdaMax[2],6000),'o',color='red')
plot(lambdaMax[3],blackbody(lambdaMax[3],7000),'o',color='green')
```

```python
from pylab import *
def calc_B(lam,T) :   
    ''' calculate blackbody radiation intensity at wavelength lambda (in nm), given temperature T in Kelvin'''    
    h = 6.626e-34    
    c = 2.99e8    
    k = 1.381e-23    
    C1 = 2.0*h*c*c    
    C2 = (h*c)/k    
#C1, C2 = 1e16, 1.44e7    
    B = C1 / (lam*1.0e-9)**5 / ( exp(C2/(lam*1.0e-9)/T) - 1.0 )    
    return B

def Wien_lam(T):    
    ''' use Wien's law to calculate peak wavelength of blackbody radiation in nm, given a temperature T in Kelvin '''    
    lambda_max = ((0.0028978) / T )/(1.0e-9)    
    return lambda_max

lam = linspace(100.0,1000.0,100) 
# wavelengths
rcParams.update({'font.size': 24})
fig1 = figure(figsize=(10,8))

for T in (4000,5000,6000,7000) :    
    # extra credit :    
    if T == 4000 :        
        Tcol = 'red' # colorization   
    elif T == 5000 :       
        Tcol = 'orange'   
    elif T == 6000 :       
        Tcol = 'green'   
    elif T == 7000 :      
        Tcol = 'blue'
    B = calc_B(lam,T)   
    lam_max = Wien_lam(T)  
    B_max = calc_B(lam_max,T)    
    plot(lam,B,color=Tcol,label=str(T)+' K')    
    plot(lam_max,B_max,'o',color=Tcol)
    legend(loc='best',fontsize=21)
    xlabel('$\lambda \, \mathrm{[nm]}$')
    #ylabel('$B(\lambda, T)$')
    ylabel('$B(\lambda, T) \, \, \mathrm{[W/m^3/sterad]}$')
    title('Blackbody Radiation')
    
grid('on')
show()
#savefig('Midterm_Fig.png')
```
