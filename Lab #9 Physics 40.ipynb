```python
from pylab import *

def height() :
    height=zeros(len(t))
    i = 0
    while i < 14 :
        z_f = 9 - (1/2)*9.8*(t[i]**2)
        height[i] = z_f
        print('The height is %.3f meters at time %.1f seconds' % (height[i],t[i]))
        i+=1
        if i == 14:
            t[i]=1.355261854 #This time is where the height is equal to 0
            z_f = 9 - (1/2)*9.8*(t[i]**2)
            height[i] = z_f
            print('The height is %.3f meters at time %.3f seconds' % (height[i],t[i]))
    return height

t=arange(0,1.5,.1)
H=height()

fig1 = figure(figsize=(12,12));
title('Ball Height Versus Time', fontsize=25)
xlabel('$Time$ (seconds)', fontsize=25)
ylabel('$Height$ (meters)', fontsize=25)
grid(color='black', linestyle='-', linewidth=.5)
rcParams['figure.figsize'] = 12,12
rcParams.update({'font.size': 20})
xticks(arange(0.0, 1.5, step=0.1))
yticks(arange(0, 10, step=1))
plot(t,H, color='red', linewidth=2)
#savefig('Henry_Wahhab_Lab9_DisplacementGraph.png', bbox_inches = 'tight')

print('') #space between outputs

def velocity() :
    velocity=zeros(len(t))
    i = 0
    while i <= 14 :
            if i==0:
                v_f = 0
                velocity[i] = v_f
                print('The velocity is %.3f m/s at time %.1f seconds' % (velocity[i],t[i]))
                i+=1
            elif i < 14 :
                v_f = (H[i]-H[i-1])/(t[i]-t[i-1])
                velocity[i] = v_f
                print('The velocity is %.3f m/s at time %.1f seconds' % (velocity[i],t[i]))
                i+=1
            elif i==14 :
                H[i]=0
                t[i]=1.355261854 #Time where height is equal to 0
                v_f = (H[i]-H[i-1])/(t[i]-t[i-1])
                velocity[i] = v_f
                print('The velocity is %.3f m/s at time %.3f seconds' % (velocity[i],t[i]))
                break
            else:
                break
    return velocity

S=velocity()

fig2 = figure(figsize=(12,12));
title('Ball Velocity Versus Time', fontsize=25)
xlabel('$Time$ (seconds)', fontsize=25)
ylabel('$Velocity$ (m/s)', fontsize=25)
grid(color='black', linestyle='-', linewidth=.5)
rcParams['figure.figsize'] = 12,12
rcParams.update({'font.size': 20})
xticks(arange(0.0, 1.5, step=0.1))
yticks(arange(0, -15, step=-1))
plot(t,S,color='blue',linewidth=2,label='dz/dt')
plot(t,-9.8*t,color='orange',linewidth=2,label='-gt')
legend(loc='best',fontsize=30)
#savefig('Henry_Wahhab_Lab9_VelocityGraph.png', bbox_inches = 'tight')

#The difference between the two functions for velocity is due to subintervals causing the dz/dt estimation to simply be average velocity
#between two points instead of instantaneous velocity at a given time which is calculated with -gt (v_f=v_i+a*t)
```
