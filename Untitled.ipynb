```python
from pylab import *

x=array([1,4,7,10,13])
y=array([2,4,6,7,12])

fig1 = figure(figsize=(10,7));
for c in range(len(x)):
    plot(x[c],y[c],'o',markersize=5)

def Slope():
    Slope=zeros(len(x))
    for i in range(len(x)):
        Slope[i]=y[i]/x[i]
    return Slope

S=Slope()

def Slope2():
    Slope2=zeros(len(x))
    for i in range(len(x)):
        if i==1:
            Slope2[i]=y[i]/x[i]
        Slope2[i]=(y[i]-y[i-1])/(x[i]-x[i-1])
    return Slope2

S2=Slope2()

AS=sum(S)/len(x) #Average Slope
AS2=sum(S2)/len(x)

x1=linspace(x[0],x[len(x)-1],50)
y1=AS*x1
y2=AS2*x1
y3=(AS+AS2)*x1/2
plot(x1,y1,color='red',label='Average Slope each Point')
plot(x1,y2,color='green',label='Between Points')
plot(x1,y3,color='magenta',label='Double Avg')
legend(loc='best',fontsize=20)
```

```python
from numpy import *

num1=int(input('Enter first number: '))
sign=input('Enter mathematical operation: ')
num2=int(input('Enter second number: '))

def calculator(num1, sign, num2):
    if sign == '+':
        calculator=(num1 + num2)
        print(num1, '+', num2, '=', calculator)
    elif sign == '-' :
        calculator=(num1 - num2)
        print(num1, '-', num2, '=', calculator)
    elif sign == '*':
        calculator=(num1 * num2)
        print(num1, '*', num2, '=', calculator)
    elif sign == '/':
        calculator=(num1 / num2)
        print(num1, '/', num2, '=', calculator)
    elif sign == '**':
        calculator=(num1 ** num2)
        print(num1, '**', num2, '=', calculator)
    elif sign == '//':
        calculator=(num1 // num2)
        print(num1, '//', num2, '=', calculator)
    elif sign == '%':
        calculator=(num1 % num2)
        print(num1, '%', num2, '=', calculator)
    else:
        calculator=nan
        print('Invalid operation/number')
    return calculator
    
CC=calculator(num1,sign,num2)
```
